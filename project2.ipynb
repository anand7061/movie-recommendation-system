```python
import nltk
#this library is used in to convert (for eg. loving to love,loved to love,love to love)
# it will convert in sigle word from different similar words 
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))

    return " ".join(y)
        
```

```python
# for eg. 
print(ps.stem('dancing'))
print(ps.stem('danced'))
print(ps.stem('dance'))
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title']== movie].index[0]
    distances=similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
        
 
```
